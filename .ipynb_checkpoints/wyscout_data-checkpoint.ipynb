```python
#importing necessary libraries
import pathlib
import os
import pandas as pd
import json
```

## Competition data

```python
path = 'wyscout_data/competitions.json'

#open data
with open(path) as f:
    data = json.load(f)
    #save it in dataframe

df_competitions = pd.DataFrame(data)
#structure of data
df_competitions.info()
```
