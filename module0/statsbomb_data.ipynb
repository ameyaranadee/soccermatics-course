```python
from mplsoccer import Sbopen
# The first thing we have to do is open the data. We use a parser SBopen available in mplsoccer.
parser = Sbopen()
```

```python
parser
```

## Competition data

```python
#opening data using competition method
df_competition = parser.competition()
#structure of data
df_competition.info()
```

```python
df_competition.head()
```

### Champions League data

```python
df_competition[df_competition['competition_name'] == 'Champions League'].sort_values(by=['season_name'])
```

### UCL Final 2018/19

```python
#opening data using match method
df_match = parser.match(competition_id=16, season_id=4)
#structure of data
df_match.info()
```

```python
df_match['match_id']
```

```python
#opening data using match method
df_lineup = parser.lineup(22912)
#structure of data
df_lineup.info()
```

### Teams based on line-up

```python
spurs_df = df_lineup[df_lineup['team_name'] == 'Tottenham Hotspur']
liverpool_df = df_lineup[df_lineup['team_name'] == 'Liverpool']
```

## Event Data

```python
#opening data
df_event, df_related, df_freeze, df_tactics = parser.event(22912)

#structure of data
df_event.info()
```

### Extracting passing data

```python
df_passes = df_event[['player_id', 'player_name', 'position_name', 'pass_recipient_id', 'pass_recipient_name', 'pass_length', 'pass_angle', 'pass_height_id', 'pass_height_name', 'end_x', 'end_y']]
df_passes = df_passes.dropna()
```
